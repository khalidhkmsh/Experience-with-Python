## Numerical investigation of an oscillator

We consider a particle of mass $m$ moving on the horizontal $x$ axis in a potential $V (x) = \omega^2_0\frac{x^2}{2}$,
subject to a dragging force $F_{drag} = −\gamma \dot{x}$˙ and a driving force $F_{drive} = \alpha \cos(t + \phi)$. The equation of
motion is therefore of the general form
$$ \ddot{x}+\frac{\gamma}{m}\dot{x}+\omega^2_0-\frac{\alpha}{m} \cos(t + \phi)= 0. \qquad(2)$$
This is a harmonic oscillator with damping and driving. We want to study the solutions of this
equation in terms of the free parameters of the problem: the initial position $x_0$, the initial velocity $\dot{x}_0$,
the frequency $\omega_0$ and the parameters $\gamma,\alpha$ and $\phi$. Once we have the solution we want to produce plots
of the position, the velocity, the potential, kinetic and mechanical energies, and the phase portrait
(graph of the velocity as function of the position)

1. Write a Python code to solve equation (2) for a given set of data $(x_0, \dot{x}_0, \omega_0,\gamma , \alpha, \phi)$

```python
import numpy as np
from scipy.integrate import odeint
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#time
a=0
b=100
h=0.01
n=int((b-a)/h)
t=np.linspace(a,b,n+1)
```

```python
def ODE3(z,t):
    x,u=z
    dxdt=u
    dudt=((α/m)*np.cos(t+ϕ))-(γ/m)*u-(ω**2)*x
    return np.array([dxdt,dudt])
```

2. Using the solution found with your code, write a code which produces the three following
graphs:

(a) position $x(t)$ and velocity $\dot{x}(t)$ both as functions of time,

(b) kinetic energy $T(t)$, potential energy $V (t)$ and 
mechanical energy $E(t) = T(t) + V (t)$ as
functions of time,

(c) the phase portrait $\dot{x}$ as function of  $x$.

For instance, with

$
\alpha =0 ms^{-2},
\phi=0,
\gamma=0,
m=1kg,
x_0= 5m
x_0=0ms^{-1},
\omega=0.8rads^{-1}
$

Observe the conservation of energy

```python
# constants
α=0
ϕ=0
γ=0
m=1
ω=0.8
#in_cond
z0=[5,0]

# #time
# a=0
# b=60
# h=0.01
# n=int((b-a)/h)
# t=np.linspace(a,b,n+1)
# solve
z=odeint(ODE3,z0,t)
x=z[:,0]
u=z[:,1]
```

```python
T= 0.5*m*u**2
V=0.5*(ω*x)**2
E=V+T
```

```python
#plot position x(t) and velocity x ̇(t) both as functions of time,
plt.figure(figsize=(15,3))

plt.subplot(1,3,1)

plt.plot(t,x,label='x(t)')
plt.plot(t,u,label='$ \dotx $ (t)')
plt.xlabel('t (s)')
plt.legend()
# plots of energy

plt.subplot(1,3,2)
plt.plot (t, T, label="T(t)")
plt.plot (t, V, label="V(t)")
plt.plot (t, E, label="E(t)")
plt.xlabel('t (s)')
plt.legend()

# the phase portrait x ̇ as function of x.
plt.subplot(1,3,3)
plt.plot(x,u) 
x0,y0=z0[0],z0[1]
plt.plot(x0, y0, marker="o", markersize=10, markeredgecolor="green", markerfacecolor="black")
plt.xlabel("x (m)")
plt.ylabel("$\dotx$ (m/s)")

plt.legend()
plt.show()
```
