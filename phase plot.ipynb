sketch the phase potrait of 
   $$\dot{x}=
  \left[ {\begin{array}{cc}
   1 & 1 \\
   2 & -1 \\
  \end{array} } \right]x$$

```python
import numpy as np
import  matplotlib.pyplot as plt
```

```python
# ##creating a square grid

lw= -3       ## lower limit
up= 3        ## upperlimit
stp=4      ## slices in the interval

### two vectors that define the x and y axis values on the grid
x_axs= np.linspace(lw,up,stp)
y_axs= np.linspace(lw,up,stp)
```

```python
### creating the matrix of cordinstes
x, y =np.meshgrid(x_axs, y_axs) 
```

```python
#x
```

```python
#y
```

```python
dxdt= x+y
dydt= 2*x-y
# ##Draw a line through eigen vctors  
# a= np.linspace(lw ,up, stp)
# plt.plot(a, (3**0.5 -1)*a) 
# plt.plot(a, (3**0.5 +1)*(-a))
```

```python
plt.streamplot(x,y,dxdt,dydt)
```

skech phase potrait for $$\begin{cases}
\dot{x}=-2x - 6y + 8\\
\dot{y}= 8x + 4y - 12\\
\end{cases}$$

```python
X ,Y= np.meshgrid(np.linspace(-5,5,5), np.linspace(-5,5,5))
x_dt= -2*X-6*Y+8
y_dt= 8*X+4*Y-12
plt.streamplot(X, Y,x_dt,y_dt)
```

skech phase potrait for $$\begin{cases}
\dot{x}=x - y\\
\dot{y}= x^2 + y^2 - 2\\
\end{cases}$$

```python
x_V ,y_v =np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))
dt_x= x_V- y_v
dt_y= x_V**2+y_v**2-2
plt.streamplot(x_V, y_v, dt_x, dt_y)
```
